```python
import numpy as np
import matplotlib.pyplot as plt
```

## Part a:

```python
## I will take all the constants as mentioned below
## I0 = 100, alpha_nu0 = 1, R0 = 1, j_nu0 = 1
```

```python
## Functions to solve the equation
def first_part(theta):
    I0 = 100
    alpha_0 = 1
    R0 = 1
    return I0*np.exp(  -alpha_0* (np.cos(theta))**2 * R0**2 / 2)

def second_part(theta):
    alpha_0 = 1
    R0 = 1
    j_nu0 = 1
    multiplier = j_nu0*(np.cos(theta))**2

    h = 1e-3        ##Stepsize
    r = np.arange(1e-5, R0, h)
    integral = np.sum(np.exp(-alpha_0*(np.cos(theta))[:,None]**2 * r[None,:]**2 / 2)/r[None,:], axis=1) * h

    return multiplier*integral


## Solution
theta = np.arange(0, 2*np.pi + 1e-2, 1e-2)
I = first_part(theta=theta) + second_part(theta=theta)

## Plotting the result
plt.figure(figsize = (12, 12))
ax = plt.subplot(111, projection='polar')
ax.plot(theta, first_part(theta=theta), color = 'r', lw = 2, label = "Contribution by the source")
ax.plot(theta, second_part(theta=theta), color = 'g', lw = 2, label = "Contribution by the medium")
ax.plot(theta, I, color='b', linewidth=2, label = "Intencity")
ax.set_title("Variation of intensity with angle at the point r=0")
ax.grid(True)
plt.legend()
plt.show()
```

## Part b:

```python
## I will take all the constants as mentioned below
## I0 = 1, alpha_nu0 = 1, R0 = 1, j_nu0 = 1
```

```python
## Functions to solve the equation

def first_part_b(theta):

    I0 = 1
    alpha_0 = 1
    R0 = 1

    h = 1e-3
    r = np.arange(0, R0, h)

    tau = np.sum(alpha_0*r*np.exp(-r/R0))*h

    return I0*np.exp(-tau) * np.ones_like(theta)


def second_part_b(theta):

    alpha_0 = 1
    R0 = 1
    j0 = 1

    h = 1e-3
    r = np.arange(1e-5, R0, h)

    tau_r = np.cumsum(alpha_0*r*np.exp(-r/R0))*h

    integrand = np.exp(-tau_r)

    integral = np.sum(integrand)*h

    multiplier = j0*np.cos(theta)*np.sin(theta)

    return multiplier*integral


## Solution

theta = np.arange(0,2*np.pi+1e-2,1e-2)

I = first_part_b(theta) + second_part_b(theta)


## Plotting

plt.figure(figsize=(12,12))
ax = plt.subplot(111,projection='polar')

ax.plot(theta, first_part_b(theta), color='r', lw=2, label="Contribution by the source")
ax.plot(theta, second_part_b(theta), color='g', lw=2, label="Contribution by the medium")
ax.plot(theta, I, color='b', lw=2, label="Intensity")

ax.set_title("Variation of intensity with angle at r=0 (Case b)")
ax.grid(True)
plt.legend()

plt.show()
```

```python
## Functions to solve the equation

def first_part_c(theta):

    I0 = 1
    alpha_0 = 1
    R0 = 1

    h = 1e-3
    r = np.arange(1e-5, R0, h)

    tau = np.sum(alpha_0*r*np.log(r/R0))*h

    return I0*np.exp(-tau)*np.ones_like(theta)


def second_part_c(theta):

    alpha_0 = 1
    R0 = 1
    j0 = 1

    h = 1e-3
    r = np.arange(1e-5, R0, h)

    tau_r = np.cumsum(alpha_0*r*np.log(r/R0))*h

    integrand = r*np.exp(-tau_r)

    integral = np.sum(integrand)*h

    return j0*integral*np.ones_like(theta)


## Solution

theta = np.arange(0,2*np.pi+1e-2,1e-2)

I = first_part_c(theta) + second_part_c(theta)


## Plot

plt.figure(figsize=(12,12))
ax = plt.subplot(111,projection='polar')

ax.plot(theta, first_part_c(theta), color='r', lw=2, label="Contribution by the source")
ax.plot(theta, second_part_c(theta), color='g', lw=2, label="Contribution by the medium")
ax.plot(theta, I, color='b', lw=2, label="Intensity")

ax.set_title("Variation of intensity with angle at r=0 (Case c)")
ax.grid(True)
plt.legend()
plt.show()
```
