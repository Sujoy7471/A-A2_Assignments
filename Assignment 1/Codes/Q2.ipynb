```python
import numpy as np
import matplotlib.pyplot as plt
```

I am taking alpha_nu0 = 1 and R0 = 1

```python
## Part a:
def tau_a(r):
    return np.pi*(1 - r**2)/2

## Part b:
def tau_b(r):
    return 2*np.pi*( (1+r)*np.exp(-r) - 2*np.exp(-1) )

## Part c:
def tau_c(r):
    
    tau_vals = []

    for ri in r:
        x = np.arange(ri, 1, 1e-4)
        integrand = x*np.log(x)
        tau = np.sum(integrand)*1e-4
        tau_vals.append(2*np.pi*tau)

    return np.array(tau_vals)
```

```python
r = np.arange(1e-5, 1, 1e-4)
```

```python
plt.figure(figsize = (12, 8))
plt.plot(r,tau_a(r), label = "Optical depth for the part a")
plt.plot(r,tau_b(r), label = "Optical depth for the part b")
plt.plot(r,tau_c(r), label = "Optical depth for the part c")
plt.xlabel("r")
plt.ylabel("Optical depth")
plt.gca().invert_xaxis()
plt.grid()
plt.legend()
plt.show()
```
