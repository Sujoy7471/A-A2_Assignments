```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Temperature range
T = np.arange(300,1001,100)

# radial range
r = np.linspace(0.1,5,100)

# constants
sigma = 5.67e-8
alpha0 = 1
sigma0 = 1
R0 = 1
dTdr = 1

fig,ax = plt.subplots(3,1,figsize=(8,10))

# -------- case (a) --------
for temp in T:
    Fa = (16*sigma*temp**3)/(3*(alpha0+sigma0)*r**2)*abs(dTdr)
    ax[0].plot(r,Fa,label=f"T = {temp} K")

ax[0].set_title("Case (a) : $\\alpha \\propto r^2$")
ax[0].set_xlabel("r")
ax[0].set_ylabel(r"$F(r)=\frac{16\sigma T^3}{3(\alpha_0+\sigma_0)r^2}\left|\frac{\partial T}{\partial r}\right|$")

ax[0].grid(True,which="major")
ax[0].minorticks_on()
ax[0].grid(True,which="minor",linestyle=":")

ax[0].legend()


# -------- case (b) --------
for temp in T:
    Fb = (16*sigma*temp**3)/(3*(alpha0+sigma0)*r)*abs(dTdr)
    ax[1].plot(r,Fb,label=f"T = {temp} K")

ax[1].set_title("Case (b) : $\\alpha \\propto r$")
ax[1].set_xlabel("r")
ax[1].set_ylabel(r"$F(r)=\frac{16\sigma T^3}{3(\alpha_0+\sigma_0)r}\left|\frac{\partial T}{\partial r}\right|$")
ax[1].grid(True,which="major")
ax[1].minorticks_on()
ax[1].grid(True,which="minor",linestyle=":")

ax[1].legend()


# -------- case (c) --------
for temp in T:
    Fc = (16*sigma*temp**3)/(3*(alpha0+sigma0)*np.exp(-r/R0))*abs(dTdr)
    ax[2].plot(r,Fc,label=f"T = {temp} K")

ax[2].set_title("Case (c) : $\\alpha \\propto e^{-r/R_0}$")
ax[2].set_xlabel("r")
ax[2].set_ylabel(r"$F(r)=\frac{16\sigma T^3}{3(\alpha_0+\sigma_0)e^{-r/R_0}}\left|\frac{\partial T}{\partial r}\right|$")
ax[2].grid(True,which="major")
ax[2].minorticks_on()
ax[2].grid(True,which="minor",linestyle=":")

ax[2].legend()

plt.tight_layout()
plt.show()
```
